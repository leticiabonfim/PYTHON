getElementById( - getElementById() é um comando (método) do JavaScript usado para encontrar e acessar um elemento HTML na página pelo seu atributo id.

document.getElementById("botao-curtir") pega o elemento com id "botao-curtir" na página para você poder manipular ele com JavaScript.

12 - (MTI31201PR25) Em um sistema de estacionamento, o valor cobrado depende do número
de horas que o carro ficou estacionado. A cada hora completa, é cobrado R$ 5,00, e o valor
total é calculado somente se o tempo for maior que zero. O código a seguir verifica o tempo
estacionado e retorna o valor a ser pago:
function calculaEstacionamento(horasTotais) {
if (horasTotais > 0) {
let valorTotal = horasTotais * 5;
return `Total a pagar: R$ ${valorTotal}`;
} else {
return "Nenhum valor a pagar.";
}
}
Ao chamar a função com o parâmetro horasTotais sendo 3, qual será a saída do código?
A) Nenhum valor a pagar
B) Total a pagar: R$ 15,00
C) Total a pagar: R$ 20,00
D) Total a pagar: R$ 25,00
E) Total a pagar: R$ 10,00


#→ Palavras que podem dificultar entendimento da resolução:

return `Total a pagar: R$ - Usado dentro de uma função para enviar um valor de volta para quem chamou essa função.
Quando o return é executado, a função termina e o valor indicado é entregue.

function - Palavra-chave para criar uma função em JavaScript.
Funções são blocos de código que realizam tarefas específicas e podem ser reutilizadas.
else - Parte do controle condicional (junto com if).
Serve para executar um bloco de código quando a condição do if não é verdadeira.
